## Variable Passing

```python
%%writefile var_pass.c

#include <stdio.h>

void change_value(int a) {
    a = 5;
}

void change_value_ref(int *a) {
    *a = 10;
}

void swap(int *x, int *y) {

   int temp;
   temp = *x;    /* save the value at address x */
   *x = *y;      /* put y into x */
   *y = temp;    /* put temp into y */
   return;
}

int main() {
    int x = 0;
    change_value(x); // value doesn't change
    printf("%d\n", x);
    change_value_ref(&x); // value changes
    printf("%d\n", x);

    int a = 10;
    int b = 20;
    printf("a, b = %d %d \n", a, b);
    swap(&a, &b);
    printf("After swap: a, b = %d %d \n", a, b);
}
```

```python
%%writefile isprime.c

#include <stdio.h>

int is_prime(int n){
    for (int i = 2; i*i <= n; i++)
    {
        if (n%i==0)
        {
            return 0;
        }
    }
    return 1;
}


int main(){
    int a;
    scanf("%d", &a);
    if (is_prime(a))
    {
        printf("%d is prime",a);
    } else {
        printf("%d is not prime", a);
    }

}
```

```python
%%writefile prime_sieve.c

#include <stdio.h>




void prime(int n)
{
    // make a list of number from 2 to n
    int nos[n - 1];
    int isprime[n - 1]; // a list showing if ith number in nos is prime or not 1 for prime 0 else
    for (int i = 0; i < n - 1; i++)
    {
        nos[i] = 2 + i;
        isprime[i] = 1;
    }


    int prime_index = 0;
    int prime_now = nos[prime_index];

    // first prime

    while (prime_now * prime_now < n)
    {
        int index = prime_index + 1;
        while (isprime[index] == 0)
        {
            index++;
        }

        while (index < n - 1)
        {
            if(nos[index]%prime_now ==0) {isprime[index]=0;}
                index += 1;
                while (isprime[index] == 0)
                {
                    index++;
                }
        }
        prime_index += 1;
        while (isprime[prime_index] == 0)
        {
            prime_index++;
        }

        prime_now = nos[prime_index];
    }

    for (int i = 0; i < n - 1; i++)
    {
        if (isprime[i])
        {
            printf("%d, ", nos[i]);
        }
    }
}

int main()
{
    int a;
    scanf("%d", &a);
    prime(a);
}
```

## Searching and Sorting

```python
%%writefile linsearch.c

#include <stdio.h>

int linsearch(int A[], int x, int n){
    for (int i = 0; i < n; i++)
    {
        if (A[i]==x)
        {
            return i;
        }

    }
    return -1;
}


int main() {
    int A[] = {2,3,0,9, 10};
    int a;
    scanf("%d", &a);
    printf("position of %d in A is %d", a, linsearch(A, a, 5));
}
```

```python
%%writefile binsearch.c

#include <stdio.h>

int binsearch(int A[], int x, int n){
    int l = 0;
    int h = n-1;
    int m;
    while (l<=h)
    {
        m = (l+h)/2;
        if (A[m]==x)
        {
            return m;
        }
        if (A[m]> x)
        {
            h = m-1;
        } else {
            l = m+1;
        }

    }
    return -1;
}

int main(){
    int A[] = {1,3,7,8,11,15,20,25,30};
    int a;
    scanf("%d", &a);
    printf("position of %d in A is %d", a, binsearch(A,a, sizeof(A)/sizeof(A[0])));
}
```

```python
%%writefile prime_sieve.cpp

#include <iostream>
#include <vector>
//we use vector instead of array
//a vector is like a variable sized array in which we can insert and delete values

using namespace std;
/**
* The function  prime(n) prints a list of primes up to n.
* Adopts the Sieve of Eratosthenes algorithm
* removes the non-prime numbers from the sieve
* a prime is able to divide the number.
*/
void primes_upto_n(int n){
        vector<int> primes ;
        // create a seive from 2 to n
        for (int i = 0; i < n-1; i++){
                primes.push_back(i + 2);
        }
        int prime_index = 0;
        int index;
        int prime_now = primes[prime_index];
        while (prime_now * prime_now < n) {
                index = prime_index + 1;
                while ( index <= primes.size()-1) {
                        if (primes[index] % prime_now != 0) {
                                index ++;
                        } else {
                                primes.erase(primes.begin() + index);
                        }
                }
                prime_index++;
                prime_now = primes[prime_index];
        }

        for (int i = 0; i < primes.size(); i++) {
                cout << primes[i] << " " ;
        }
        cout << endl;
}

int main(){
        int n;
        cout << "Enter the number" << endl;
        cin >> n;
        primes_upto_n(n);
        return 0;
}
```

```python
%%writefile binsearch.cpp

#include <iostream>

using namespace std;

int binsearch(int A[], int x, int n){
    int l = 0;
    int h = n-1;
    int m;
    while (l<=h)
    {
        m = (l+h)/2;
        if (A[m]==x)
        {
            return m;
        }
        if (A[m]> x)
        {
            h = m-1;
        } else {
            l = m+1;
        }

    }
    return -1;
}

int main(){
    int A[] = {1,3,7,8,11,15,20,25,30};
    int a;
    cin >> a;
    cout << "position of " << a << " in A is " << binsearch(A,a, sizeof(A)/sizeof(A[0]));
}
```
