# Print code with line nos

```python
code = """
import multiprocessing
import numpy as np
import random
import time

max_procs = 4; N = 10**9
np.random.seed(0)
a = np.random.random(N)
b = np.random.random(N)
datasize = N//max_procs

def local_array_mult_sum(proc):
  init_ind = proc*datasize
  end_ind = (proc+1)*datasize
  return np.sum(a[init_ind:end_ind]
    * b[init_ind:end_ind])

if __name__ == '__main__':
  start = time.time()
  wall_start = time.perf_counter()
  cpu_start = time.process_time()
  inputs = list(range(max_procs))

  with multiprocessing.Pool(max_procs) as pool:
    local_sum = pool.map(local_array_mult_sum,
      inputs)

  wall_end = time.perf_counter()
  cpu_end = time.process_time()
  end = time.time()
  pll_sum = sum(local_sum)
  print(f"Parallel sum using {max_procs}
    processors")
  print(f"Total_sum  = {pll_sum}")
  print(f"Time using time.time()={end-start:.6f}")
  print(f"Wall-Clock Time=
    {wall_end-wall_start:.6f}")
  print(f"CPU time taken=
    {cpu_end - cpu_start:.6f} \n")
  wall_start = time.perf_counter()
  cpu_start = time.process_time()
  sequential_sum = np.sum(a*b)
  wall_end = time.perf_counter()
  cpu_end = time.process_time()
  print(f"Sequential sum using 1 processor")
  print(f"Total_sum  = {sequential_sum}")
  print(f"Wall-Clock Time =
    {wall_end - wall_start:.6f}")
  print(f"CPU time taken=
    {cpu_end - cpu_start:.6f}")
  print(f"Are the pll and sequential nos close?
    : {isclose(pll_sum, sequential_sum)}")
"""

for i, line in enumerate(code.strip().split('\n'), 1):
    print(f"{i:02d} | {line}")
```
